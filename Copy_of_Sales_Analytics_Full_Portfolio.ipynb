## Dataset Overview

```python


The dataset for this sales analytics project consists of synthetic sales data generated to simulate real-world transactions. It includes 1000 records across 7 columns, providing a foundational structure for sales analysis. The key columns include:

*   **OrderDate:** Timestamp of each sales order.
*   **ProductID:** Unique identifier for each product.
*   **ProductName:** Name of the product sold (e.g., 'Laptop', 'Mouse', 'Keyboard').
*   **Region:** Geographic region where the sale occurred (e.g., 'North', 'South', 'East', 'West').
*   **Quantity:** Number of units sold in a particular order.
*   **UnitPrice:** Price per unit of the product.
*   **TotalPrice:** Calculated total price for the order (Quantity * UnitPrice).

Initial inspection revealed no missing values or duplicate records, making the dataset clean and ready for immediate analysis. Time-based features such as OrderYear, OrderMonth, OrderDay, OrderDayOfWeek, and OrderHour were engineered to facilitate trend analysis.
```

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Data Loading

```python


# For demonstration, let's create a synthetic sales dataset.
# In a real scenario, you would load your data from a CSV, database, etc.

num_records = 1000
dates = pd.to_datetime(pd.date_range(start='2023-01-01', periods=num_records, freq='h')) # Changed 'H' to 'h' to resolve FutureWarning
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam']
regions = ['North', 'South', 'East', 'West']

data = {
    'OrderDate': np.random.choice(dates, num_records),
    'ProductID': np.random.randint(100, 105, num_records),
    'ProductName': np.random.choice(products, num_records),
    'Region': np.random.choice(regions, num_records),
    'Quantity': np.random.randint(1, 5, num_records),
    'UnitPrice': np.round(np.random.uniform(10, 1000, num_records), 2)
}

df = pd.DataFrame(data)
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

print("Synthetic sales data created. Displaying the first 5 rows:")
display(df.head())
```

## Data Inspection

```python


print("DataFrame Info:")
df.info()

print("\nDataFrame Description:")
display(df.describe(include='all'))

print("\nNumber of unique products:", df['ProductName'].nunique())
print("Number of unique regions:", df['Region'].nunique())
```

## Data Cleaning

```python


Data cleaning was performed to ensure the quality and integrity of the dataset. This involved:

1.  **Missing Value Analysis:** Checked for and confirmed the absence of any missing values across all columns.
2.  **Duplicate Analysis:** Identified and confirmed no duplicate rows were present in the dataset.

The synthetic dataset was found to be clean, requiring no further imputation or removal of erroneous entries, which ensures the reliability of subsequent analyses.
```

## Missing Value Analysis

```python


print("Missing values per column:")
print(df.isnull().sum())

# Visualize missing values (optional, for more complex datasets)
# import seaborn as sns
# import matplotlib.pyplot as plt
# plt.figure(figsize=(10, 6))
# sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
# plt.title('Missing Values Heatmap')
# plt.show()
```

## Duplicate Analysis

```python


num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")

if num_duplicates > 0:
    print("Removing duplicate rows...")
    df.drop_duplicates(inplace=True)
    print(f"Number of rows after removing duplicates: {len(df)}")
else:
    print("No duplicate rows found.")
```

## Feature Engineering

```python


# Extract time-based features from OrderDate
df['OrderYear'] = df['OrderDate'].dt.year
df['OrderMonth'] = df['OrderDate'].dt.month
df['OrderDay'] = df['OrderDate'].dt.day
df['OrderDayOfWeek'] = df['OrderDate'].dt.dayofweek # Monday=0, Sunday=6
df['OrderHour'] = df['OrderDate'].dt.hour

print("New time-based features added. Displaying head with new columns:")
display(df.head())
```

## Exploratory Data Analysis

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

print("--- Distribution of TotalPrice ---")
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPrice'], bins=30, kde=True)
plt.title('Distribution of TotalPrice')
plt.xlabel('TotalPrice')
plt.ylabel('Frequency')
plt.show()

print("\n--- Sales by Product Name ---")
plt.figure(figsize=(12, 7))
sns.barplot(x=df['ProductName'].value_counts().index, y=df['ProductName'].value_counts().values)
plt.title('Sales Count by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Number of Sales')
plt.show()

print("\n--- Total Sales by Region ---")
plt.figure(figsize=(10, 6))
sales_by_region = df.groupby('Region')['TotalPrice'].sum().sort_values(ascending=False)
sns.barplot(x=sales_by_region.index, y=sales_by_region.values, palette='viridis')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.show()
```

## KPI Development

```python


print("--- Key Performance Indicators (KPIs) ---")

total_sales = df['TotalPrice'].sum()
print(f"1. Total Sales: ${total_sales:,.2f}")

average_order_value = df['TotalPrice'].mean()
print(f"2. Average Order Value: ${average_order_value:,.2f}")

total_quantity_sold = df['Quantity'].sum()
print(f"3. Total Quantity Sold: {total_quantity_sold:,}")

num_unique_products = df['ProductID'].nunique()
print(f"4. Number of Unique Products Sold: {num_unique_products}")

top_selling_product = df.groupby('ProductName')['Quantity'].sum().idxmax()
top_selling_product_qty = df.groupby('ProductName')['Quantity'].sum().max()
print(f"5. Top Selling Product by Quantity: {top_selling_product} ({top_selling_product_qty} units)")

highest_revenue_product = df.groupby('ProductName')['TotalPrice'].sum().idxmax()
highest_revenue_product_revenue = df.groupby('ProductName')['TotalPrice'].sum().max()
print(f"6. Product with Highest Revenue: {highest_revenue_product} (${highest_revenue_product_revenue:,.2f})")


top_region_by_sales = df.groupby('Region')['TotalPrice'].sum().idxmax()
top_region_sales_value = df.groupby('Region')['TotalPrice'].sum().max()
print(f"7. Top Region by Sales: {top_region_by_sales} (${top_region_sales_value:,.2f})")
```

## Trend Analysis

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("--- Monthly Sales Trend ---")
monthly_sales = df.groupby('OrderMonth')['TotalPrice'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values)
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

print("\n--- Daily Sales Trend (Day of Week) ---")
daily_sales = df.groupby('OrderDayOfWeek')['TotalPrice'].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=daily_sales.index, y=daily_sales.values, hue=daily_sales.index, palette='plasma', legend=False) # Fixed FutureWarning
plt.title('Total Sales by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Sales ($)')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("--- Monthly Sales Trend for TotalRevenue ---")
monthly_sales_revenue = df.groupby('OrderMonth')['TotalPrice'].sum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales_revenue.index, y=monthly_sales_revenue.values, marker='o')
plt.title('Total Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue ($)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

## Category Analysis

```python


import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("--- Sales Performance by Product Category ---")
product_sales = df.groupby('ProductName').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalRevenue=('TotalPrice', 'sum')
).sort_values(by='TotalRevenue', ascending=False)

print("\nProduct Sales Summary:")
display(product_sales)

# Plot Total Revenue by Product Name
plt.figure(figsize=(12, 7))
sns.barplot(x=product_sales.index, y=product_sales['TotalRevenue'], hue=product_sales.index, palette='coolwarm', legend=False) # Fixed FutureWarning
plt.title('Total Revenue by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue ($)')
plt.show()

# Plot Total Quantity Sold by Product Name
plt.figure(figsize=(12, 7))
sns.barplot(x=product_sales.index, y=product_sales['TotalQuantity'], hue=product_sales.index, palette='coolwarm', legend=False) # Fixed FutureWarning
plt.title('Total Quantity Sold by Product Name')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Sold')
plt.show()
```

### Regional Customer Segmentation Analysis

Since individual customer IDs are not available in our synthetic dataset, we will segment customers based on the `Region` they belong to. We will then analyze key metrics for each region to understand their purchasing patterns and identify potential differences.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("--- Customer Segment Analysis by Region ---")

regional_customer_analysis = df.groupby('Region').agg(
    TotalOrders=('OrderDate', 'count'),
    TotalQuantitySold=('Quantity', 'sum'),
    TotalRevenue=('TotalPrice', 'sum'),
    AverageOrderValue=('TotalPrice', 'mean')
).sort_values(by='TotalRevenue', ascending=False)

print("\nRegional Customer Segments Summary:")
display(regional_customer_analysis)

# Visualize Average Order Value by Region
plt.figure(figsize=(10, 6))
sns.barplot(x=regional_customer_analysis.index, y=regional_customer_analysis['AverageOrderValue'], hue=regional_customer_analysis.index, palette='viridis', legend=False)
plt.title('Average Order Value by Region')
plt.xlabel('Region')
plt.ylabel('Average Order Value ($)')
plt.show()

# Visualize Total Quantity Sold by Region
plt.figure(figsize=(10, 6))
sns.barplot(x=regional_customer_analysis.index, y=regional_customer_analysis['TotalQuantitySold'], hue=regional_customer_analysis.index, palette='plasma', legend=False)
plt.title('Total Quantity Sold by Region')
plt.xlabel('Region')
plt.ylabel('Total Quantity Sold')
plt.show()
```
